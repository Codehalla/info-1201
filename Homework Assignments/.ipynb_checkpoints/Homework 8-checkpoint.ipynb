```python
# Problem 1
def calculateAcceptanceScore(studentData):
    sat = (float(studentData[0])/160)*0.15
    gpa = (float(studentData[1])*2)*0.15
    curriculum = float(studentData[2])*0.1
    interest = float(studentData[3])*0.1
    if (studentData[4] == True):
        donor = 10*0.5
    else:
        donor = 0
    math = sat + gpa + curriculum + interest + donor
    roundUp = round(math,2)
    return roundUp
```

```python
# Problem 2
# Student,SAT,GPA,Interest,High School Quality,Sem1,Sem2,Sem3,Sem4,Donor
def calculateAcceptanceScore(studentData):
    sat = (float(studentData[1])/160)*0.15
    gpa = (float(studentData[2])*2)*0.15
    interest = float(studentData[3])*0.1
    quality = float(studentData[4])*0.1
    if (studentData[9] == 'TRUE'):
        donor = 10*0.5
    else:
        donor = 0
    acceptanceScore = sat + gpa + interest + quality + donor
    acceptanceScoreRound = round(acceptanceScore, 2)
    return acceptanceScoreRound


def createAcceptanceDictionary(fileName):
    import csv
    studentScores = {}
    
    file = open(fileName, 'r')
    csvReader = csv.reader(file, delimiter=",")
    next(csvReader)
    
    for row in csvReader:
        studentScores[row[0]] = float(calculateAcceptanceScore(row))
    return studentScores
    
    file.close()
```

```python
# Problem 4
# def calculateAcceptanceScore 
    # Calculates score with donors included
def calculateAcceptanceScore(studentData):
    sat = (float(studentData[1])/160)*0.15
    gpa = (float(studentData[2])*2)*0.15
    interest = float(studentData[3])*0.1
    quality = float(studentData[4])*0.1
    if (studentData[9] == 'TRUE'):
        donor = 10*0.5
    else:
        donor = 0
    acceptanceScore = sat + gpa + interest + quality + donor
    acceptanceScoreRound = round(acceptanceScore, 2)
    return acceptanceScoreRound

def calculateAcceptanceScoreSansDonor(studentData):
    sat = (float(studentData[1])/160)*0.3
    gpa = (float(studentData[2])*2)*0.3
    interest = float(studentData[3])*0.2
    quality = float(studentData[4])*0.2
    acceptanceScore = sat + gpa + interest + quality
    acceptanceScoreRound = round(acceptanceScore, 2)
    return acceptanceScoreRound

# Main function to call --> def getScores
def getScores(fileName, studentName):
    import csv
    studentScores = {}
    studentScoresSansDonors = {}
    studentScoreComparison = {}
    
    # Open file and use csvReader
    file = open(fileName, 'r')
    csvReader = csv.reader(file, delimiter=",")
    next(csvReader)
    
    for row in csvReader:
        # Creating a dictionary for student score with donor factor
        studentScores[row[0]] = float(calculateAcceptanceScore(row))
        
        # creating a dictionary for student score without donor factor
        studentScoresSansDonors[row[0]] = float(calculateAcceptanceScoreSansDonor(row))
        
    x = studentScores[studentName]
    y = studentScoresSansDonors[studentName]
    studentScoreComparison['Score excluding donor info'] = y
    studentScoreComparison['Score including donor info'] = x
    return studentScoreComparison

    
    file.close()
```

```python
getScores('ProspectiveStudentData.csv', 'Betty Sumitro')
```
