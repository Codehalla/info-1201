```python
# let's write a function given an artist name, will print out each album of that band the function
# will take two parameters: the name of the file that contains the album information
# and the name of the artist we are looking for

def printBandAlbums(fileName, artistName):
    #  open file
    myFile = open(fileName, 'r')
    # we're going to skip the header
    myFile.readline()
    # we'll look at each line in the file
    for line in myFile:
        # split the line up using a delimiter
        lineParts = line.split(',')
        # if the data on the line is for the artist we are looking for,
        # we will print some information out that album
        # lineParts[3] is the artist
        if (lineParts[3] == artistName):
            # this is the artist we are looking for
            # let's print album name, the rank, and genre
            print("\nAlbum Name: " + lineParts[2] + "\nRank: " + lineParts[0] + "\nGenre: " +  lineParts[4])
            
    # close  file
    myFile.close()
```

```python
def printBandAlbumsFromList(fileName, artistName):
    # now let's store all the albums for a particular artist in a list
    albumList = []  # create an empty list to store the album names
    myFile = open(fileName, 'r')   # open file
    # skip header
    myFile.readline()
    allAlbumList = myFile.read().splitlines()
    # we read the entire text of fileName in using .read() --> this creates a string
    # then we split that string into lines using .splitlines()
    # .splitlines() creates a list using the newline character as a delimiter
    for album in allAlbumList:
        # split the line up
        albumParts = album.split(",")
        # is this the artist we are looking for?
        if (albumParts[3] == artistName):
            # store the album name in our list
            albumList.append(albumParts[2])

    # now, print the number of albums and the names of them
    
    if (len(albumList) == 0):
        print(artistName + " has no albums in the RS TOP 500")
    else:
        print(artistName + " has " + str(len(albumList)) + " albums in the RS TOP 500")
        for album in albumList:
            print(album)
    
    # close  file
    myFile.close()
```

```python
printBandAlbumsFromList("RollingStoneAlbumList.csv", "David Bowie")
```

```python
# We ran into  problems when data had commas in it
# let's leverage some exisiting code that will help us handle this situations this existing
# code is the csv module

# a module is code that adds functionality to 'core' python
    #  a module is like a container for additional functions
```

```python
# we're going to import the csv modile to use the stuff in it
import csv
```

```python
myFile = open("RollingStoneAlbumList.csv", 'r')
csvReader = csv.reader(myFile, delimiter = ',', quotechar = '"')
# this syntax is a little different because we specify parameters (with variableName= )
# because these parameters are 'optional'. These parameters have default settings in 
# the funciton definition.

# The first parameter in the reader function is te name of the file handle
# the second parameter is the delimiter
# the third parameter is the character that is used to treat everything inside as text
    # for example, if you have "Live, At the Apollo", the comma is not treated as a delimiter

# Now we're going to use csvReader to iterate through the csv file

header = next(csvReader) # this grabs the next line inside the csvReader and 
# assigns it to the variable number 
print(header)
line = next(csvReader)
print(line)

# next() grabs the next line and returns the data split on the delimiter character in a list
# it essentially acts like .readline followed by .split(',')
# furthermore, any delimiter inside the quotechars is treated like text, not a delimiter

myFile.close()
```

```python
# we can process whole csv file much easier with the csv header
myFile = open('RollingStoneAlbumList.csv', 'r')
csvReader = csv.reader(myFile, delimiter=",", quotechar='"')
header = next(csvReader)
for row in csvReader:
    print(row[2] + " by " + row[3])
myFile.close()
```

```python
def printArtistAlbumInfo(fileName, artistName):
    print('Artist Name: ' + artistName)
    # open file
    myFile = open(fileName, 'r')
    
    # create counter
    counter = 0
    
    # create csv reader
    csvReader = csv.reader(myFile, delimiter=",", quotechar='"')
    # skip header
    next(csvReader)
    
    # read the remaining rows
    for row in csvReader:
        if (artistName == row[3]):
            print(row[2] + ', Rank: ' + row[0])
            counter += 1
    print('There are ' + str(counter) + ' albums')
    # close file
    myFile.close()
```

```python
printArtistAlbumInfo('RollingStoneAlbumList.csv', 'James Brown')
```
