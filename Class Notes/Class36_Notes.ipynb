```python
# Looking at some unstructured text and counting up the number of times each word in the text is used
# We will write a function so we can use this with any text file

def countAllWords(fileName):
    # open file
    wordsFile = open(fileName,"r").read().split()
    
    # let's get all the words and store them
    # read everything in at once
    # Gives us one huge string
    # splits string based on whitespace
    # print(wordsFile)
    
    # We have to look at each word, store each word, and count the number of words
    wordCounts = {}
    
    for words in wordsFile:
        # lets make words lowercase
        words = words.lower()    # .lower makes every letter in the string lowercase
        # .upper makes every letter in the string uppercase
        
        # Let's deal with punctutation
        # myString.replace(thingToReplace, thingToReplaceWith)
        # .replace will replace 'thingToReplace' with 'thingToReplaceWith' inside of myString
        
        words = words.replace(".", "")
        words = words.replace(",", "")
        words = words.replace("?", "")
        words = words.replace("!", "")
        
        
        # see if word is not in wordCounts
        if(words not in wordCounts):
            # add word to dictionary and set its value equal to 1
            wordCounts[words] = 1
        else:
            # if word is in the dictionary, we must increase its value by 1
            wordCounts[words] += 1
    return wordCounts
    wordsFile.close()
```

```python
words = countAllWords('AllBeatlesLyrics.txt')
```

```python
print(words)
# Problem: 
    # capitalized words get counted differently than lowercase
    # words with punctuation gets counted differently than words without it
    
# Solution:
    # make words all lowercase/uppercase
```

```python
import sortdictionary
sortedWords = sortdictionary.sortDictionaryByValues(words, True)
```

```python
print(sortedWords)
```

```python
# The words that we see the most often are things like pronouns and conjunctions and articles, words
# that just don't have a lot of meaning to the collection of words
# we call these words 'Stop words'
    # they are words that we want to ignore
# to see if a word is a stop word, we'll check each word and see if it's in that stop word file

# let's write a function that takes a dictionary as a parameter and a filename as a parameter where
# the fileName is the file of stopwords. The function should return a dictionary that does not
# have any stopwords in it

def removeStopWordsFromDictionary(dictionaryName, stopWordsFileName):
    # create empty dicitonary to store the non-stop words
    dictionaryNoStopWords = {}
    # open stop words file
    stopWordsFile = open(stopWordsFileName, 'r')
    # let's create a list of stop words based on what's in the file
    stopWordsList = stopWordsFile.read().split()
    stopWordsFile.close()
    
    # Let's read through each word in the dictionary that we passed into the function if the word
    # is not a stop word, copy it into dictionaryNoStopWords
    
    for word in dictionaryName:     # word is the variable storing each key one at a time
        if (word not in stopWordsList):
            # add word to the dictionary
            dictionaryNoStopWords[word] = dictionaryName[word]
            # how do we get the value from dictionary for a given key?
            # dictionaryName[keyName]
            # This creates a new entry in dictionaryNoStopWords where word is the key
            # and the value is the same value that's in dictionaryName
            # (which is the dictionary we passed into the function)
    return dictionaryNoStopWords
```

```python
noSWSorted = removeStopWordsFromDictionary(sortedWords, "stop-word-list.txt")
```
